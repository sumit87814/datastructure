```python
# printing even number using range and list
even_number=[num for num in range(21) if num%2==0]
print(even_number)

# using a empty lsit and loop
even=[]
for num in range(21):
    if num %2==0:
        even.append(num)
print(even)

# using other comprehensive way
numbers = [1, 2, 3, 4, 5]
result=[(num,'even') if num%2==0 else (num,"odd") for num in numbers]
print(result)
```
